```python
separator = '-'
y = []
for x in all_nba["year"]:
    y.append(int(x.split(separator, 1)[0]))
#result = my_str.split(separator, 1)[0]
all_nba["year"] = y
all_nba = all_nba[all_nba['year'] >= 1996]
all_nba = all_nba[all_nba['year'] <= 2017]
```

```python
all_nba["player1"] = all_nba["player1"].str.replace(" C","")
all_nba["player2"] = all_nba["player2"].str.replace(" F","")
all_nba["player3"] = all_nba["player3"].str.replace(" F","")
all_nba["player4"] = all_nba["player4"].str.replace(" G","")
all_nba["player5"] = all_nba["player5"].str.replace(" G","")
```

```python
all_nba
```

```python
df["all_nba"] = "n/a"
df["all_nba_enc"] = 4
df["all_nba_total"] = 0
for index, row in all_nba.iterrows():
    year = row["year"]
    team = row["team"]
    enc = 4
    if team == "1st":
        enc = 1
    elif team == "2nd":
        enc = 2
    elif team == "3rd":
        enc = 3
    player1 = row["player1"]
    df.loc[(df['playerName'] == player1) & (df['seasonStartYear'] == year),'all_nba'] = team
    df.loc[(df['playerName'] == player1) & (df['seasonStartYear'] == year),'all_nba_enc'] = enc
    df.loc[(df['playerName'] == player1) & (df['seasonStartYear'] >= year),'all_nba_total'] =  df.loc[(df['playerName'] == player1) & (df['seasonStartYear'] >= year),'all_nba_total'] + 1
    player2 = row["player2"]
    df.loc[(df['playerName'] == player2) & (df['seasonStartYear'] == year),'all_nba'] = team
    df.loc[(df['playerName'] == player2) & (df['seasonStartYear'] == year),'all_nba_enc'] = enc
    df.loc[(df['playerName'] == player2) & (df['seasonStartYear'] >= year),'all_nba_total'] =  df.loc[(df['playerName'] == player2) & (df['seasonStartYear'] >= year),'all_nba_total'] + 1
    player3 = row["player3"]
    df.loc[(df['playerName'] == player3) & (df['seasonStartYear'] == year),'all_nba'] = team
    df.loc[(df['playerName'] == player3) & (df['seasonStartYear'] == year),'all_nba_enc'] = enc
    df.loc[(df['playerName'] == player3) & (df['seasonStartYear'] >= year),'all_nba_total'] =  df.loc[(df['playerName'] == player3) & (df['seasonStartYear'] >= year),'all_nba_total'] + 1
    player4 = row["player4"]
    df.loc[(df['playerName'] == player4) & (df['seasonStartYear'] == year),'all_nba'] = team
    df.loc[(df['playerName'] == player4) & (df['seasonStartYear'] == year),'all_nba_enc'] = enc
    df.loc[(df['playerName'] == player4) & (df['seasonStartYear'] >= year),'all_nba_total'] =  df.loc[(df['playerName'] == player4) & (df['seasonStartYear'] >= year),'all_nba_total'] + 1
    player5 = row["player5"]
    df.loc[(df['playerName'] == player5) & (df['seasonStartYear'] == year),'all_nba'] = team
    df.loc[(df['playerName'] == player5) & (df['seasonStartYear'] == year),'all_nba_enc'] = enc
    df.loc[(df['playerName'] == player5) & (df['seasonStartYear'] >= year),'all_nba_total'] =  df.loc[(df['playerName'] == player5) & (df['seasonStartYear'] >= year),'all_nba_total'] + 1
```
