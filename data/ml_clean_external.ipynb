```python
df["mvp"] = 0
df["mvp_rank"] = 100000
df["mvp_total"] = 0

for index, row in mvp.iterrows():
    year = row["year"]
    rank = row["Rank"]
    name = row["Player"]
    df.loc[(df['seasonStartYear'] == year) & (df['playerName'] == name),'mvp_rank'] = int(rank)
    if rank == '1':
        df.loc[(df['seasonStartYear'] == year) & (df['playerName'] == name),'mvp'] = 1
        df.loc[(df['playerName'] == name) & (df['seasonStartYear'] >= year),'mvp_total'] =  df.loc[(df['playerName'] == name) & (df['seasonStartYear'] >= year),'mvp_total'] + 1
    
```

```python
week = pd.read_csv(r'C:\Users\raoma\Documents\michigan\ml\EECS-545---Predicting-NBA-Player-s-Salary\data\raw_data\NBA_player_of_the_week.csv')
week["Date"] = week["Date"].astype(str)

def convert_date(data):
    #print(type(data))
    if not isinstance(data,str):
       return np.nan
    split_height = data.split(' ')
    
    return int(split_height[2])

week["Date"] = week['Date'].apply(convert_date)
week = week[week['Date'] <= 2017]
week.head()
```

```python
df["player_week_enc"] = 0
df["player_week_total"] = 0

for index, row in week.iterrows():
    year = row["Date"]
    name = row["Player"]
    
    df.loc[(df['playerName'] == name) & (df['seasonStartYear'] == year),'player_week_enc'] =  df.loc[(df['playerName'] == name) & (df['seasonStartYear'] == year),'player_week_enc'] + 1
    df.loc[(df['playerName'] == name) & (df['seasonStartYear'] >= year),'player_week_total'] =  df.loc[(df['playerName'] == name) & (df['seasonStartYear'] >= year),'player_week_total'] + 1
```
